Questions to answer:

    1. What was the most frequent movie duration in the 1990s?

```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")
```

```python
#Analysing the first 10 rows
netflix_df.head(10)
```

```python
# Filtering for required years
cond1 = (netflix_df["release_year"] >= 1990) & (netflix_df["type"] == 'Movie')
cond2 = (netflix_df["release_year"] <= 1999) & (netflix_df["type"] == 'Movie')

# Combining the conditions
filtered_df = netflix_df[cond1 & cond2]

# Obtaining value counts for duration of movies in 1990s
duration_counts = filtered_df["duration"].value_counts()
duration = duration_counts.index[0]
print(duration)
```
